```python
class Node:
    def __init__(self, left=None, right=None, feature=None, threshold=None, value=None):
        self.feature = feature
        self.threshold = threshold
        self.left = left
        self.right = right
        self.value = value

    def is_leaf_node(self):
        return self.value is not None
```

```python
class DecisionTree:
    def __init__(self, min_sample_split=2, max_depth=100, n_features=None, information_gain_type='entropy'):
        self.min_sample_split = min_sample_split
        self.max_depth = max_depth
        self.n_features = n_features
        self.information_gain_type = information_gain_type
        
    @staticmethod
    def entropy(y, information_gain_type='entropy'):
        # count occurences of targets, normalize it
        if len(y) == 0:
            return 0
        ps = np.bincount(y[:,0]) / len(y[:,0])
        if type == 'entropy':
            return -np.sum([p * np.log2(p) for p in ps if p > 0])
        else:  # type=='gini'
            return 1 - np.sum([p ** 2 for p in ps])
```

```python
%%add_to DecisionTree
def fit(self, X, y):
        # adjust self.n_features size if features provided by X fewer than expected
        self.n_features = X.shape[1] if not self.n_features else min(self.n_features, X.shape[1])
        # grow the tree using a recursive class method
        self.root = self._grow_tree(X,y)
        
def predict(self, X):
    # traverse tree for every data row and return output as an array
    return np.array([self._traverse_tree(x, self.root) for x in X])
```

```python
%%add_to DecisionTree
def _grow_tree(self, x, y, depth=0):
    # initialize sizes based on the parameters
    n_samples, n_features = x.shape
    n_labels = len(np.unique(y)) 
    # stopping criteria:
    # max depth reached OR just one label type left OR not enough samples remaining in vector x
    if (depth >= self.max_depth or 
        n_labels < 2 or
        n_samples < self.min_sample_split
       ):
        leaf_value = self._most_common_label(y)
        # return leaf node
        pepa = Node(value=leaf_value) 
        return pepa
    # randomly choose features of size n_features from the features 
    feature_idxs = np.random.choice(n_features, self.n_features, replace=False)
    # search for best split at this node
    best_feature, best_threshold = self._find_best_criteria(x, y, feature_idxs)
    #left_idxs, right_idxs = self_split_node(x[:, best_feature], best_threshold)
    left_idxs = np.argwhere(x[:, best_feature] <= best_threshold).flatten()
    right_idxs = np.argwhere(x[:, best_feature] > best_threshold).flatten()

    # recursively continue growing
    left = self._grow_tree(x[left_idxs], y[left_idxs], depth + 1)
    right = self._grow_tree(x[right_idxs], y[right_idxs], depth + 1)
    # return decision node
    return Node(left=left, right=right, feature=best_feature, threshold=best_threshold)
```

```python
%%add_to DecisionTree
def _find_best_criteria(self, x, y, feature_idxs):
    # initialize function parameters
    max_gain = -1000
    split_threshold, split_idx = None, None
    # iterate through available feature indexes
    for idx in feature_idxs:
        # generate relevant x column for this feature
        x_col = x[:,idx]
        # generate all possible splits = all unique feature values in x
        thresholds = np.unique(x_col)
        # iterate through available thresholds
        for thd in thresholds: 
            # calculate information gain for this threshold at this feature
            information_gain = self._calculate_information_gain(thd, x_col, y)
            # is this the best information gain so far?
            if information_gain > max_gain:
                max_gain = information_gain
                split_threshold = thd
                split_idx = idx
    return split_idx, split_threshold
```

```python
%%add_to DecisionTree
def _calculate_information_gain(self, threshold, x_column, y):
    # calculate parent entropy
    parent_entropy = self.entropy(y)
    # split x_column into left and right based on threshold
    left_branch_idxs = np.argwhere(x_column <= threshold).flatten()
    right_branch_idxs = np.argwhere(x_column > threshold).flatten()
    # calculate entropy for left and right branches
    left_entropy = self.entropy(y[left_branch_idxs])
    right_entropy = self.entropy(y[right_branch_idxs])
    # calculate child entropy
    n = len(y); n_left = len(left_branch_idxs); n_right = len(right_branch_idxs)
    child_entropy = (n_left / n) * left_entropy + (n_right / n) * right_entropy
    # return information change
    return parent_entropy - child_entropy    
```

```python
%%add_to DecisionTree
@staticmethod
def _most_common_label(y):
    return np.bincount(y[:,0]).argmax()
```

```python
%%add_to DecisionTree
def _traverse_tree(self, x, node):
    if node.is_leaf_node():
        return node.value
    if x[node.feature] <= node.threshold:
        return self._traverse_tree(x, node.left)
    else:
        return self._traverse_tree(x, node.right)
```

# Custom Random Forest implementation

```python
class RandomForest:
    def __init__(self, n_trees=100, min_sample_split=2, max_depth=100, n_features=None):
        self.n_trees = n_trees
        self.min_sample_split = min_sample_split
        self.n_features = n_features
        self.max_depth = max_depth
        self.trees = []

    def fit(self, x, y):
        # make sure our tree array is empty
        self.trees = []
        # learn n_trees trees
        for i in range(self.n_trees):
            print(f'Learning... #{i+1} tree out of total #{self.n_trees} trees.')
            # init i-th tree
            tree = DecisionTree(min_sample_split=self.min_sample_split,
                                max_depth=self.max_depth, n_features=self.n_features)
            # bootstrap sub-sample of the training data
            x_sample, y_sample = self._select_sample(x,y)
            # train tree
            tree.fit(x_sample, y_sample)
            # this will make a fine addition to my (tree) collection
            self.trees.append(tree)
    
    def predict(self, x):
        # go through our trees and get its prediction for the feature vector x
        forest_predictions = np.array([tree.predict(x) for tree in self.trees]).T
        # popular vote the prediction of the random forest
        y_pred = [self._most_common_label(prediction) for prediction in forest_predictions]
        return y_pred
    
    def _select_sample(self, x, y):
        # bootstrap aggregation done here.. select a random sample from our data
        n_samples = x.shape[0]
        idxs = np.random.choice(n_samples, size=n_samples, replace=True)
        return x[idxs], y[idxs]

    def _most_common_label(self, y):
        # get most common label 
        counter = Counter(y)
        most_common = counter.most_common(1)[0][0]
        return most_common
```
